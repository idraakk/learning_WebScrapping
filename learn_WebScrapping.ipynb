# Web Scrapping https://www.worldometers.info/world-population/population-by-country/

```python
# Web Scrapping https://www.worldometers.info/world-population/population-by-country/
# making a table for country wise population
url= urlopen('https://www.worldometers.info/world-population/population-by-country/')
html=url.read()
soup=BeautifulSoup(html,'html.parser')

# saving the table
table=soup.find_all('table', { 'id' : "example2"})[0]
table_rows=table.find_all('tr')

rows_num=len(table.find_all('tr'))
cell_num=len(table.find_all('tr')[0].find_all('th'))

# storing all the headings in a list
column_heads=[]
for column in (table.find_all('tr')[0].find_all('th')):
    column_heads.append(column.text)

# storing all data in an array of dictionary
temp={}
dict=[]
for i in range(1,rows_num):
    table_row=table_rows[i].find_all('td')
    for j in range(0, cell_num):
        temp[column_heads[j]]=table_row[j].text
    dict.append(temp.copy()) # .copy() beacause in python, here temp acts as a pointer
    temp.clear()

# to export dict into a csv file
#import pandas as pd
#df = pd.DataFrame(dict)
#df.to_csv('population.csv', index=False)


# SHORT METHOD USING PANDAS
# df=pd.read_html(str(soup))[0]
# df
```
